```python
#Eliminar widgets
dbutils.widgets.removeAll()
```

```python

#Importar librerias
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql import functions as F
```

```python
# Widgets
dbutils.widgets.text("catalogo", "sanvar_kim")
dbutils.widgets.text("esquema_source", "silver")
dbutils.widgets.text("esquema_sink", "golden")
```

```python
# Variables
catalogo = dbutils.widgets.get("catalogo")
esquema_source = dbutils.widgets.get("esquema_source")
esquema_sink = dbutils.widgets.get("esquema_sink")
```

```python
# Datos a utilizar filtrados
df_golden_persona = spark.table(f"{catalogo}.{esquema_source}.detalle_alquileres") \
    .select(
        "consecutivo",
        "identificacion",
        "nombre_inquilino",
        "apellido_inquilino",
        "direccion",
        "id_apartamento",
        "habitaciones",
        "cochera",
        "servicios",
        "ubicacion"
    )
display(df_golden_persona)

```

```python
%sql

-- Ambiente para carga de datos
USE CATALOG sanvar_kim;
USE SCHEMA silver;
```

```python
#Escribir datos completos en tabla golden
df_golden_persona.write.mode("overwrite").insertInto("sanvar_kim.golden.load_alquileres")
```

```python
#Confirmar datos incertados
display(
    spark.table("sanvar_kim.golden.load_alquileres")
)
```
